1. Validar que la longitud sea 6.
2. Validar que de la posicion 0 a la 2 sean letras.
3. Validar que de la posicion 3 al final sean numeros.

```python
c = "c"
c >= 'a' and c <= 'z'
```

```python
def validar_placa(text):
    if len(text) != 6:
        return False
    for char in text[:3]:
        if char < 'a' or char > 'z':
            return False
    for char in text[3:]:
        if char < '0' or char > '9':
            return False
    return True
```

```python
x = "xyz123"
```

```python
validar_placa(x)
```

```python
x = "ab1c23"
```

```python
validar_placa(x)
```

```python
def filter_length(text, thr):
    words = text.split(" ")
    filtered_words = filter(
        lambda word: len(word) >= thr,
        words
    )
    return " ".join(filtered_words)
```

```python
print(filter_length(text, 5))
```
