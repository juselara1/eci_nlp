```python
def validate(seq: str) -> str:
    if len(seq) != 6:
        return "Invalida"
    letter_candidates = seq[:3]
    all_letters = all(map(is_letter, letter_candidates))

    num_candidates = seq[3:]
    all_nums = all(map(is_num, num_candidates))
    return "Valida" if all_letters and all_nums else "Invalida"
```

```python
for candidate in x:
    is_valid = validate(candidate)
    print(f"La placa {candidate} es {is_valid}")
```

```python
from unidecode import unidecode
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud

nltk.download("stopwords")
sw = stopwords.words("english")
```

```python
def get_nums(text: str) -> int | None:
    try:
        pat = re.compile(r"[\d\.]+")
        mat = re.search(pat, text)
        return float(mat.group(0))
    except:
        return None

def clean_text(text: str) -> str:
    spch_pat = re.compile(r"[^a-z\s]")
    spaces_pat = re.compile(r"\s+")
    lower_text = text.lower()
    norm_text = unidecode(lower_text)
    spch_text = re.sub(spch_pat, " ", norm_text)
    spaces_text = re.sub(spaces_pat, " ", spch_text)
    tokens = spaces_text.split(" ")
    filtered_words = filter(lambda token: token not in sw, tokens)
    return " ".join(filtered_words)

df_preprocess = (
    df
    .rename(columns={
        "Restaurant": "restaurant",
        "Reviewer": "reviewer",
        "Review": "review",
        "Rating": "rating",
        "Time": "time",
        "Pictures": "pictures"
        })
    .assign(
        time = lambda df: pd.to_datetime(df.time),
        reviewer_reviews = lambda df: (
            df.Metadata.str.extract(pat=r"(\d+).*\d+")
        ),
        reviewer_followers = lambda df: (
            df.Metadata.str.extract(pat=r"\d+.*(\d+)")
        ),
        rating = lambda df: df.rating.apply(get_nums)
    )
    .drop(columns=["Metadata", "7514"])
    .dropna()
    .astype({
        "restaurant": "string",
        "reviewer": "string",
        "review": "string",
        "reviewer_reviews": "int",
        "reviewer_followers": "int"
    })
    .assign(
        review = lambda df: df.review.apply(clean_text),
        rating = lambda df: df.rating.astype("int")
    )
    .reset_index(drop=True)
)
df_preprocess.info()
```

```python
groups = (
    df_preprocess
    .groupby("rating")
    .apply(lambda group: " ".join(group.review))
)
```

```python
fig, axes = plt.subplots(5, 1, figsize=(7, 25))
for i in range(1, 6):
    ax = axes[i - 1]
    text = groups[i]
    wc = WordCloud(background_color="white").generate_from_text(text)
    ax.imshow(wc)
    ax.axis("off")
    ax.set(title=f"Rating: {i}")
```
