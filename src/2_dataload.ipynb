```python
data = [
    ("Juan Lara", 27),
    ("Sergio", 25),
    ("Anny", 21),
    ("Andres", 21)
]
```

```python
data_str = ["nombre,edad"]
for row in data:
    data_str.append(f"{row[0]},{row[1]}")
data_str = "\n".join(data_str)
```

```python
print(data_str)
```

```python
f = open("edades.txt", mode="w")
f.write(data_str)
f.close()
```

```python
f = open("edades.txt", mode="a")
f.write("\nPepe,27")
f.close()
```

```python
f = open("edades.txt", mode="r+")
print(f.read())
f.write("hola mundo")
f.close()
```

```python
f = open("edades.txt", mode="w+")
f.write("hola mundo")
f.seek(0)
print(f.read())
f.close()
```

```python
f = open("data.txt", mode="wb")
f.write(b"Hola")
f.close()
```

```python
img = mpimg.imread("poema.jpg")
```
